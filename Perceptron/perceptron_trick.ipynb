```python
from sklearn.datasets import make_classification
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=42,hypercube=False,class_sep=2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[y==0][:,0],X[y==0][:,1],color='red',label='Class 0')
plt.scatter(X[y==1][:,0],X[y==1][:,1],color='blue',label='Class 1')
```

```python
def step(x):
    return 1 if x>=0 else 0
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    w=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,X.shape[0])
        y_hat=step(np.dot(X[j],w))
        w=w+lr*(y[j]-y_hat)*X[j]
    return w[0],w[1:]                           
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print("Intercept:",intercept_)
print("Coefficients:",coef_)
```

```python
m=-coef_[0]/coef_[1]
b=-intercept_/coef_[1]
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
plt.scatter(X[y==0][:,0],X[y==0][:,1],color='red',label='Class 0')
plt.scatter(X[y==1][:,0],X[y==1][:,1],color='blue',label='Class 1')
plt.plot(x_input,y_input,color='green',label='Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```
